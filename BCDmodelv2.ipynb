# Data Preprocessing
1: Encoding categorical data
2: Scaling Data

```python
#Encodeing Categorical Data:
le = LabelEncoder()
data.diagnosis = le.fit_transform(data.diagnosis)
```

```python
data.head()
```

```python
#Removing unnecessary columns and splitting data into X with variables and y with only outcome.

X = data.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1)
y = data['diagnosis']
```

```python
X.head()
```

```python
y.head()
```

```python
X.describe()
```

```python
#Scaling the Data
from sklearn.preprocessing import StandardScaler
standardizer = StandardScaler()
X_1 = standardizer.fit_transform(X)
```

```python
#Spliting Data into Train and Test Datasets:
X_train, X_test, y_train, y_test = train_test_split(X_1, y, test_size=0.2, random_state=42)
```

For this Dataset, I will be performing the ML algorithm SVM to classify each diagnosis as either malignant or benign.

```python
# Importing SVM
from sklearn.svm import SVC

svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
```

```python
print("Score on Test: " + str(svm_model.score(X_test, y_test)*100))
print("Score on Train: "+ str(svm_model.score(X_train, y_train)*100))
```

Based on the above score, one can see how the model distinguishes at each point what constitutes a malignant or benigne diagnosis. The overall accuracy of the ML testing was approximately 95.6% on the test size, and 98.7% on the train set, respectfully, at the time this model was run. Thank you.
